```python
import re
import numpy as np

MAX_DIG  = 4
MAX_IND  = 256
MAX_VAL  = MAX_IND * MAX_IND - 1

def hex_4(d):
    h = format(d, 'X').zfill(MAX_DIG)
    return h[:2] + ':' + h[2:]

def generate_chromosome(p):
    size = int(p * MAX_VAL)
    syn = np.random.choice(
        MAX_VAL, size=size, replace=False
    )
    syn = list(map(hex_4, syn))
    return sorted(syn)
```

```python
p1 = np.log(MAX_IND) / MAX_IND + np.random.normal() / 100
p2 = np.log(MAX_IND) / MAX_IND + np.random.normal() / 100

c1 = generate_chromosome(p1)
c2 = generate_chromosome(p2)

print(c1)
print('')
print(c2)
```

```python
print(len(c1))
print(len(c2))
```

```python
matches = []
for i, syn in enumerate(c1):
    if syn in c2:
        matches += [(i, c2.index(syn))]
matches
```

```python
match = matches[np.random.randint(0, len(matches))]
match
```

```python
print(c1[match[0]], c2[match[1]])
```

```python
c3 = c1
c1_x = c1[:match[0]] + c2[match[1]:]
c2_x = c2[:match[1]] + c3[match[0]:]

print(';'.join(c1_x))
print('')
print(';'.join(c2_x))
```
